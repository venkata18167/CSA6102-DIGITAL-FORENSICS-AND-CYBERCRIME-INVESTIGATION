```python

with open("storage.bin", "wb") as f:
    f.write(b"RANDOMDATA")
    f.write(b"%PDF-1.4 Deleted PDF File")
    f.write(b"XXXX")
    f.write(b"\x89PNG\r\n\x1a\n Deleted PNG File")
    f.write(b"YYYY")
    f.write(b"\xFF\xD8\xFF Deleted JPG File")
signatures = {
    b"%PDF": "PDF Document",
    b"\x89PNG\r\n\x1A\n": "PNG Image",
    b"\xFF\xD8\xFF": "JPEG Image"
}
with open("storage.bin", "rb") as file:
    data = file.read()

print("Scanning storage for deleted files...\n")

found = False

for signature, file_type in signatures.items():
    position = data.find(signature)
    if position != -1:
        print(file_type, "found at byte", position)
        found = True

if not found:
    print("No deleted files found.")
```
