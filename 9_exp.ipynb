```python

with open("login_log.txt", "w") as file:
    file.write("192.168.1.10 FAIL\n")
    file.write("192.168.1.10 FAIL\n")
    file.write("192.168.1.10 FAIL\n")
    file.write("192.168.1.20 SUCCESS\n")
    file.write("192.168.1.30 FAIL\n")
    file.write("192.168.1.30 FAIL\n")
    file.write("192.168.1.40 SUCCESS\n")
    file.write("192.168.1.30 FAIL\n")

failed_attempts = {}

with open("login_log.txt", "r") as file:
    for line in file:
        ip, status = line.strip().split()

        if status == "FAIL":
            failed_attempts[ip] = failed_attempts.get(ip, 0) + 1

print("Brute-Force Detection Report\n")

found = False

for ip, count in failed_attempts.items():
    if count >= 3:
        print("Suspicious IP :", ip)
        print("Failed Attempts :", count)
        print("Status : Brute-Force Attack Detected\n")
        found = True

if not found:
    print("No brute-force attack detected.")
```
