```python

with open("email_header.txt", "w") as file:
    file.write("From: attacker@gmail.com\n")
    file.write("Return-Path: fake@yahoo.com\n")
    file.write("Received: from 192.168.1.100\n")
    file.write("Subject: Urgent Account Verification\n")

headers = {}

with open("email_header.txt", "r") as file:
    for line in file:
        if ":" in line:
            key, value = line.split(":", 1)
            headers[key.strip()] = value.strip()

print("Email Header Analysis\n")

print("From        :", headers.get("From"))
print("Return-Path :", headers.get("Return-Path"))
print("Received    :", headers.get("Received"))

# Detect spoofing
if headers.get("From") != headers.get("Return-Path"):
    print("\nWarning: Possible Email Spoofing Detected!")
else:
    print("\nNo Spoofing Detected.")
```
