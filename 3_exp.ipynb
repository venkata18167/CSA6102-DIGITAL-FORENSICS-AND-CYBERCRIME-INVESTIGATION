```python
# Phishing Email Detection Program

# Input email details
sender = input("Enter Sender Email: ")
subject = input("Enter Email Subject: ")
body = input("Enter Email Body: ")

# Suspicious indicators
suspicious_senders = ["fakebank.com", "unknown.com", "phishing.com"]
suspicious_keywords = [
    "urgent", "verify", "password", "click here",
    "reward", "lottery", "bank", "account"
]

# Convert to lowercase
sender = sender.lower()
subject = subject.lower()
body = body.lower()

phishing = False

# Check sender
for domain in suspicious_senders:
    if domain in sender:
        phishing = True

# Check subject and body
for word in suspicious_keywords:
    if word in subject or word in body:
        phishing = True

# Display result
if phishing:
    print("\n⚠️ Phishing Email Detected!")
else:
    print("\n✅ Email Appears Safe.")
```
