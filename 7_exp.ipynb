```python

with open("sample_pdf.pdf", "wb") as f:
    f.write(b"%PDF-1.4\nThis is a sample PDF file.")
with open("sample_png.png", "wb") as f:
    f.write(b"\x89PNG\r\n\x1a\nThis is a sample PNG file.")
with open("sample_jpg.jpg", "wb") as f:
    f.write(b"\xFF\xD8\xFF\xE0This is a sample JPG file.")

magic_numbers = {
    b"\xFF\xD8\xFF": "JPEG Image",
    b"\x89PNG\r\n\x1A\n": "PNG Image",
    b"%PDF": "PDF Document",
    b"PK\x03\x04": "ZIP Archive",
    b"GIF87a": "GIF Image",
    b"GIF89a": "GIF Image"
}

filename = input("Enter file name: ")

try:
    with open(filename, "rb") as file:
        header = file.read(8)

    file_type = "Unknown File Type"

    for signature, name in magic_numbers.items():
        if header.startswith(signature):
            file_type = name
            break

    print("File Name      :", filename)
    print("Detected Type  :", file_type)
    print("Magic Number   :", header.hex().upper())

except FileNotFoundError:
    print("File not found!")
```
