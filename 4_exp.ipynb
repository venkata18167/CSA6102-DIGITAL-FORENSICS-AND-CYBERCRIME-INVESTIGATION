```python
import socket

def dns_lookup(domain):
    try:
        print("\nDNS Lookup Results")
        print("-" * 30)


        ip = socket.gethostbyname(domain)
        print("Domain Name :", domain)
        print("IP Address  :", ip)


        host_info = socket.gethostbyaddr(ip)
        print("Host Name   :", host_info[0])


        print("\nAll IP Addresses:")
        for addr in socket.gethostbyname_ex(domain)[2]:
            print(addr)

    except socket.gaierror:
        print("DNS Lookup Failed! Invalid Domain Name.")


domain = input("Enter Domain Name: ")
dns_lookup(domain)
```
