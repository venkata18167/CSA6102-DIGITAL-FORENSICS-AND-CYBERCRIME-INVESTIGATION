```python
import shutil
import hashlib
import os

# Create a sample evidence file
with open("sample.txt", "w") as file:
    file.write("Digital Forensic Evidence File\n")
    file.write("Case ID: DF001\n")
    file.write("Investigator: T Venkat Reddy\n")
    file.write("Status: Original Evidence\n")

# Function to calculate SHA-256 hash
def calculate_hash(filename):
    sha256 = hashlib.sha256()
    with open(filename, "rb") as file:
        while True:
            data = file.read(4096)
            if not data:
                break
            sha256.update(data)
    return sha256.hexdigest()

# Original and copied file names
original_file = "sample.txt"
copied_file = "forensic_copy_sample.txt"

# Create forensic copy
shutil.copy2(original_file, copied_file)

# Calculate hashes
original_hash = calculate_hash(original_file)
copied_hash = calculate_hash(copied_file)

# Display report
print("=" * 50)
print(" DIGITAL FORENSIC BIT-STREAM COPY REPORT")
print("=" * 50)
print("Original File :", original_file)
print("Copied File   :", copied_file)

print("\nOriginal File SHA-256 Hash:")
print(original_hash)

print("\nCopied File SHA-256 Hash:")
print(copied_hash)

print("\nIntegrity Verification:")
if original_hash == copied_hash:
    print("SUCCESS: Hashes Match")
    print("The forensic copy is an exact bit-stream copy.")
else:
    print("FAILED: Hashes Do Not Match")
    print("The forensic copy has been modified.")

print("=" * 50)
```
