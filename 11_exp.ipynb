```python

with open("network_log.txt", "w") as file:
    file.write("192.168.1.10 80\n")
    file.write("192.168.1.10 443\n")
    file.write("192.168.1.10 22\n")
    file.write("192.168.1.20 80\n")
    file.write("192.168.1.20 80\n")
    file.write("192.168.1.30 21\n")
    file.write("192.168.1.30 22\n")
    file.write("192.168.1.30 23\n")
    file.write("192.168.1.30 25\n")

ip_ports = {}

# Read the network log
with open("network_log.txt", "r") as file:
    for line in file:
        ip, port = line.strip().split()

        if ip not in ip_ports:
            ip_ports[ip] = set()

        ip_ports[ip].add(port)

print("Port Scan Detection Report\n")

found = False

for ip, ports in ip_ports.items():
    if len(ports) >= 3:
        print("Suspicious IP :", ip)
        print("Ports Accessed :", ", ".join(sorted(ports)))
        print("Status : Port Scanning Detected\n")
        found = True

if not found:
    print("No port scanning detected.")
```
