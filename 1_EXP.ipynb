```python
import os
import time
from datetime import datetime

with open("evidence.txt", "w") as f:
    f.write("Original Evidence")

print("Original MTIME:",
      datetime.fromtimestamp(os.path.getmtime("evidence.txt")))

time.sleep(5)

with open("evidence.txt", "a") as f:
    f.write("\nNew Evidence Added")

print("Updated MTIME :",
      datetime.fromtimestamp(os.path.getmtime("evidence.txt")))

print("\nLocard's Exchange Principle:")
print("Every contact leaves a trace.")
```
